```python
import warnings
warnings.filterwarnings(action='ignore')
import requests
```

```python
targetSite ='https://www.starbucks.co.kr/store/getSidoList.do'
req = requests.post(targetSite)
# request 모듈의 json() 메소드로 서버에 요청
starbucks = req.json()
print(type(starbucks)) # <class 'dict'>
#print(starbucks)
sido_list = {}
for sido in starbucks['list']:
    sido_list[sido['sido_cd']] = sido.get('sido_nm')
print(sido_list)
```

```python
sido_cd = input('구군코드를 검색하려는 시도코드를 입력하세요: ')
targetSite ='https://www.starbucks.co.kr/store/getGugunList.do'
req = requests.post(targetSite, data= {
    'sido_cd': sido_cd
})

gugunList = req.json()
# print(type(gugunList))
# print(len(gugunList))
# print(gugunList.keys())
# print(gugunList)

# 구군코의 개수만큼 반복ㅏ며 구군코드('gugun_cd')를 key로 하고 구군이름('gugun_nm')을 value로 하는 딕셔너리를 만든다. 
gugun_list = {}
for gugun in gugunList['list']:
    gugun_list[gugun['gugun_cd']] = gugun['gugun_nm']
print(gugun_list)
```
