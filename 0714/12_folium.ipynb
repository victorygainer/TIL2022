```python
# folium 라이브러리를 사용하면 지도를 출력하고 지도위에 marker를 표시할 수 있다.
# folium 라이브러리를 설치하고 import한다. 
# !pip install folium
import folium
```

```python
foliumMap = folium.Map(location=[37.5662952, 126.9779451], zoom_start=16)

# folium 라이브러리의 Marker() 메소드의 location 속성으로 지정한 위치에 popup 속성으로 지정한 문자열을 표시하는 marker를 만든다.
# 작성한 marker는 add_to() 메소드를 사용해서 folium 라이브러리로 생성한 지도에 추가한다.
popup = folium.Popup('서울특별시청', max_width=300)
popup1 = folium.Popup('병수의회사', max_width=300)

folium.Marker(location=[37.5662952, 126.9779451], popup=popup).add_to(foliumMap)
folium.Marker(location=[37.400977,127.1042391], popup=popup1).add_to(foliumMap)

foliumMap


```
