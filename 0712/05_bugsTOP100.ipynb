```python
import warnings
warnings.filterwarnings(action='ignore')
import requests
from bs4 import BeautifulSoup
import datetime
```

```python
targetSite = 'https://music.bugs.co.kr/chart/track/realtime/total?wl_ref=M_contents_03_01'
req = requests.get(targetSite)
#print(req.text)
html = req.text
soup = BeautifulSoup(html, 'html.parser')
dt_now = datetime.datetime.now()
d_today = datetime.date.today()
```

```python
titles = soup.findAll('p', {'class' : 'title'})
#print(titles[0].text)
for i in range(len(titles)):
    title = titles[i].text.strip().split('\n')[0]
    print('{}위: {}'.format(i + 1, title))
```

```python
artists = soup.findAll('p', {'class' : 'artist'})
dt_now = datetime.datetime.now()
# print(artists[0].text)
print(dt_now)
for i in range(len(artists)):
    for j in range(len(titles)):
        title = titles[i].text.strip().split('\n')[0]
    # ===== for j 
    artist = artists[i].text.strip().split('\n')[0]

    print('{}위 {} - {}'.format(i + 1, title, artist))
```

```python
# 크롤링한 결과를 텍스트 파일에 저장한다. 
# file = open('/Users/jeongseung-gu/Documents/kookgi/python/venv/workspace_PythonCrawling/output/bugsTOP100.txt', 'wt') # 절대경로
filename = './output/bugsTOP100' + '_' + dt_now.strftime('%Y%m%d%H%M%S') + '.txt'
file = open(filename, 'w', -1, 'utf-8')
file.write(dt_now.strftime('%Y-%m-%d %H:%M:%S') + '\n')
for i in range(100):
    artist = artists[i].text.strip().split('\n')[0]
    title = titles[i].text.strip().split('\n')[0]
    file.write('{:3d}위 {} - {} \n'.format(i + 1, artist, title))
file.close()
print(filename[9:] + '파일로 쓰기 완료')

    
#file = open('./output/bugsTOP100.txt', 'w', -1, 'utf-8') #상대경로
```
