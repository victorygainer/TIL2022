```python
class BinaryTree:
    # 생성자에서 이진 트리의 노드를 만든다.
    def __init__(self, data = None):
        self.left = None # 왼쪽 자식 노드를 기억한다.
        self.data = data # 데이터
        self.right = None # 오른쪽 자식 노드를 기억한다.
        
    # 트리에 노드를 삽입하는 함수
    def insert(self, data):
        print('트리에 삽입하려는 데이터 {}의 부모 노드의 데이터는 {}입니다.'.format(data, self.data))
        
        # 트리에 삽입할 데이터와 부모 노드의 데이터를 비교해서 트리에 데이터가 삽입되는 위치를 결정한다.
        # 트리에 삽입할 데이터가 부모 노드의 데이터보다 작으면 부모 노드의 왼쪽에 삽입한다.
        # 트리에 삽입할 데이터가 부모 노드의 데이터보다 크면 부모 노드의 오른쪽에 삽입한다.
        
        # 트리에 삽입할 데이터가 부모 노드의 데이터보다 작은가?
        if data < self.data:
            # 부모 노드의 왼쪽에 삽입한다.
            print('부모 노드의 데이터가 삽입할 데이터보다 크기 때문에 부모 노드의 왼쪽에 삽입한다.')
            # 부모 노드의 왼쪽 링크(self.left)가 비어(None)있어야 데이터를 삽입할 수 있다.
            if self.left is None:
                # 부모 노드의 왼쪽 링크가 비어있으므로 데이터를 추가한다.
                print('부모({}) 왼쪽에 데이터 {} 추가 가능.'.format(self.data, data))
                # 새 데이터를 추가해야 하므로 트리에 추가할 데이터를 BinaryTree 클래스 객체를 생성해서 부모 노드의
                # 비어있는 왼쪽 링크에 생성된 BinaryTree 클래스 객체의 주소를 넣어준다.
                self.left = BinaryTree(data)
                print('부모({}) 왼쪽에 데이터 {} 추가 완료.'.format(self.data, data))
            else:
                # 부모 노드의 왼쪽 링크가 비어있지 않기 때문에 데이터를 추가할 수 없다.
                print('부모({}) 왼쪽에 {} 추가가 불가능하므로 부모 왼쪽 노드 {}에서 insert() 메소드를 실행한다.'.format(
                    self.data, data, self.left.data
                ))
                self.left.insert(data)
       
        # 트리에 삽입할 데이터가 부모 노드의 데이터보다 큰가?
        elif data > self.data:
            # 부모 노드의 오른쪽에 삽입한다.
            print('부모 노드의 데이터가 삽입할 데이터보다 작기 때문에 부모 노드의 오른쪽에 삽입한다.')
            # 부모 노드의 오른쪽 링크(self.right)가 비어(None)있어야 데이터를 삽입할 수 있다.
            if self.right is None:
                # 부모 노드의 오른쪽 링크가 비어있으므로 데이터를 추가한다.
                print('부모({}) 오른쪽에 데이터 {} 추가 가능.'.format(self.data, data))
                # 새 데이터를 추가해야 하므로 트리에 추가할 데이터를 BinaryTree 클래스 객체를 생성해서 부모 노드의
                # 비어있는 오른쪽 링크에 생성된 BinaryTree 클래스 객체의 주소를 넣어준다.
                self.right = BinaryTree(data)
                print('부모({}) 오른쪽에 데이터 {} 추가 완료.'.format(self.data, data))
            else:
                # 부모 노드의 오른쪽 링크가 비어있지 않기 때문에 데이터를 추가할 수 없다.
                print('부모({}) 오른쪽에 {} 추가가 불가능하므로 부모 오른쪽 노드 {}에서 insert() 메소드를 실행한다.' \
                      .format(self.data, data, self.right.data))
                self.right.insert(data)

        # 트리에 삽입할 데이터가 트리에 존재하는가?
        else:
            print('{}는(은) 트리에 존재하는 데이터입니다.'.format(data))
            
    # 이진 트리를 구성하는 노드 목록을 PreOrder 방식으로 탐색하는 함수
    # 12 => 6 => 3 => 10 => 20 => 15 => 25
    def preorder(self):
        # link(self.left, self.right)에 주소가 저장되어있으면 True, 없으면(None) False
        # 부모를 출력한다.
        print(self.data, end=' ')
        # 왼쪽 자식 노드가 있나 검사해서 있으면 preorder() 함수를 실행한다.
        if self.left:
            self.left.preorder()
        # 오른쪽 자식 노드가 있나 검사해서 있으면 preorder() 함수를 실행한다.
        if self.right:
            self.right.preorder()
            
    # 이진 트리를 구성하는 노드 목록을 InOrder 방식으로 탐색하는 함수
    # 3 => 6 => 10 => 12 => 15 => 20 => 25
    def inorder(self):
        if self.left:
            self.left.inorder()
        print(self.data, end=' ')
        if self.right:
            self.right.inorder()
    
    # 이진 트리를 구성하는 노드 목록을 PostOrder 방식으로 탐색하는 함수
    # 3 => 10 => 6 => 15 => 25 => 20 => 12
    def postorder(self):
        if self.left:
            self.left.postorder()
        if self.right:
            self.right.postorder()
        print(self.data, end=' ')
```

```python
# 이진 트리의 최상위(root) 노드를 만든다.
root = BinaryTree(12)
# 이진 트리에 데이터(노드)를 추가한다.
root.insert(12)
print('=' * 80)
root.insert(6)
print('=' * 80)
root.insert(20)
print('=' * 80)
root.insert(3)
print('=' * 80)
root.insert(10)
print('=' * 80)
root.insert(15)
print('=' * 80)
root.insert(25)
print('=' * 80)
```
