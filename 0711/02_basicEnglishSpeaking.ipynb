```python
import requests
from bs4 import BeautifulSoup
```

```python
targetSite = 'https://basicenglishspeaking.com/daily-english-conversation-topics/'
```

```python
req = requests.get(targetSite)
#print(req)
```

```python
html = req.text
#print(html)
soup = BeautifulSoup(html, 'html.parser')
#print(soup)
```

```python
subjects = [] # 파싱한 대화 주제를 저장할 빈 리스트 
```

```python
# 최근 공지사항의 제목은 class 속성이 'tail'인 <td> 태그 내부에 있다. 
divs = soup.findAll('div', {'class': 'tcb-col'})
#print(notices)
for div in divs:
    #print(div.text)
    chapter = div.findAll('a')
    #print(len(chapter))
    for a in chapter:
        #print(a.text)
        subjects.append(a.text)
```

```python
for i in range(len(subjects)):
    print('{}. {}'.format(i + 1, subjects[i]))
```

```python
def basicEnglishSpeakingMethod() :
    targetSite = 'https://basicenglishspeaking.com/daily-english-conversation-topics/'
    req = requests.get(targetSite)
    html = req.text
    soup = BeautifulSoup(html, 'html.parser')
    divs = soup.findAll('div', {'class': 'tcb-col'})
    for div in divs:
        chapter = div.findAll('a')
        for a in chapter:
            subjects.append(a.text)
    for i in range(len(subjects)):
        print('{}. {}'.format(i + 1, subjects[i]))
```

```python
basicEnglishSpeakingMethod()
```
